```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取CSV文件
data = pd.read_csv('D:\\code\\dase_introduction\\users_combined_info_500.csv')

# 查看数据的前几行
print("数据的前几行:")
print(data.head())

# 1. 国家和地区分布
country_distribution = data['country'].value_counts()

# 打印前10个主要的国家和地区
print("\n前10个主要的国家和地区:")
print(country_distribution.head(10))

# 绘制国家和地区分布图
plt.figure(figsize=(12, 6))
sns.barplot(x=country_distribution.head(10).values, y=country_distribution.head(10).index)
plt.title('国家和地区分布')
plt.xlabel('用户数量')
plt.ylabel('国家/地区')
plt.show()

# 2. 城市级别分布
city_distribution = data['location'].value_counts()

# 打印前10个主要的城市
print("\n前10个主要的城市:")
print(city_distribution.head(10))

# 绘制城市分布图
plt.figure(figsize=(12, 6))
sns.barplot(x=city_distribution.head(10).values, y=city_distribution.head(10).index)
plt.title('城市级别分布')
plt.xlabel('用户数量')
plt.ylabel('城市')
plt.show()

# 3. 事件类型分布
event_type_distribution = data['event_type'].value_counts()

# 打印前10个主要的事件类型
print("\n前10个主要的事件类型:")
print(event_type_distribution.head(10))

# 绘制事件类型分布图
plt.figure(figsize=(12, 6))
sns.barplot(x=event_type_distribution.head(10).values, y=event_type_distribution.head(10).index)
plt.title('事件类型分布')
plt.xlabel('用户数量')
plt.ylabel('事件类型')
plt.show()

# 4. 事件动作分布
event_action_distribution = data['event_action'].value_counts()

# 打印前10个主要的事件动作
print("\n前10个主要的事件动作:")
print(event_action_distribution.head(10))

# 绘制事件动作分布图
plt.figure(figsize=(12, 6))
sns.barplot(x=event_action_distribution.head(10).values, y=event_action_distribution.head(10).index)
plt.title('事件动作分布')
plt.xlabel('用户数量')
plt.ylabel('事件动作')
plt.show()

# 5. 提交频率
# 假设 'total_influence' 可以代表提交次数
commit_frequency = data['total_influence'].value_counts()

# 打印提交次数最多的前10个用户
print("\n提交次数最多的前10个用户:")
print(commit_frequency.head(10))

# 绘制提交频率图
plt.figure(figsize=(12, 6))
sns.barplot(x=commit_frequency.head(10).values, y=commit_frequency.head(10).index)
plt.title('提交频率')
plt.xlabel('用户数量')
plt.ylabel('提交次数')
plt.show()

# 6. 时间分布
# 将 'event_time' 转换为日期时间格式
data['event_time'] = pd.to_datetime(data['event_time'])

# 提取小时、日期和月份
data['hour'] = data['event_time'].dt.hour
data['date'] = data['event_time'].dt.date
data['month'] = data['event_time'].dt.month

# 6.1 按小时分布
hour_distribution = data['hour'].value_counts().sort_index()

# 打印按小时分布
print("\n按小时分布:")
print(hour_distribution)

# 绘制按小时分布图
plt.figure(figsize=(12, 6))
sns.lineplot(x=hour_distribution.index, y=hour_distribution.values)
plt.title('按小时分布')
plt.xlabel('小时')
plt.ylabel('用户数量')
plt.show()

# 6.2 按日期分布
date_distribution = data['date'].value_counts().sort_index()

# 打印按日期分布
print("\n按日期分布:")
print(date_distribution.head(10))

# 绘制按日期分布图
plt.figure(figsize=(12, 6))
sns.lineplot(x=date_distribution.index, y=date_distribution.values)
plt.title('按日期分布')
plt.xlabel('日期')
plt.ylabel('用户数量')
plt.xticks(rotation=45)
plt.show()

# 6.3 按月份分布
month_distribution = data['month'].value_counts().sort_index()

# 打印按月份分布
print("\n按月份分布:")
print(month_distribution)

# 绘制按月份分布图
plt.figure(figsize=(12, 6))
sns.barplot(x=month_distribution.index, y=month_distribution.values)
plt.title('按月份分布')
plt.xlabel('月份')
plt.ylabel('用户数量')
plt.show()
```
