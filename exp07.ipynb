```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 打印当前工作目录
print(f'当前工作目录: {os.getcwd()}')

# 指定文件的绝对路径
file_path = r'D:\newcode\dase_introduction\github_bot_processed_data.csv'

# 检查文件是否存在
if not os.path.exists(file_path):
    print(f'文件 {file_path} 不存在，请检查文件路径。')
else:
    # 读取原始数据
    raw_data = pd.read_csv(file_path)

    # 1. 删除重复数据，并输出去重前后的数据量
    print(f'去重前的数据量: {len(raw_data)}')
    data = raw_data.drop_duplicates()
    print(f'去重后的数据量: {len(data)}')

    # 2. 缺失值处理

    # 2.1 去掉 gravatar_id 列（如果存在）
    if 'gravatar_id' in data.columns:
        data = data.drop(columns=['gravatar_id'])

    # 2.2 查看各列的缺失值情况
    missing_values = data.isnull().sum()
    print("去重后各列的缺失值情况:")
    print(missing_values)

    # 2.3 将可转化成 boolean 变量的列字段转化成 boolean 变量
    boolean_columns = ['site_admin', 'hireable']
    data[boolean_columns] = data[boolean_columns].fillna(False).astype(bool)

    # 2.4 文本数据用空字符串填充空值
    text_columns = ['company', 'blog', 'location', 'bio']
    data[text_columns] = data[text_columns].fillna('')

    # 2.5 数值数据用中位数或均值填充空值
    numeric_columns = ['public_repos', 'public_gists', 'followers', 'following', 'text_bot_count']
    for col in numeric_columns:
        data[col] = data[col].fillna(data[col].median())

    # 2.6 再次查看各列有无缺失值
    missing_values_after = data.isnull().sum()
    print("处理后各列的缺失值情况:")
    print(missing_values_after)

    # 3. 数据变换，将 created_at、updated_at 转为时间戳
    data['created_at'] = pd.to_datetime(data['created_at'])
    data['updated_at'] = pd.to_datetime(data['updated_at'])
    data['created_at_timestamp'] = data['created_at'].astype(int) / 10**9
    data['updated_at_timestamp'] = data['updated_at'].astype(int) / 10**9

    # 4. 数据可视化

    # 4.1 可视化 bot 和 human 类型的情况
    type_counts = data['label'].value_counts()
    plt.figure(figsize=(8, 6))
    sns.barplot(x=type_counts.index, y=type_counts.values)
    plt.title('Bot 和 Human 类型的数量分布')
    plt.xlabel('类型')
    plt.ylabel('数量')
    plt.show()

    # 报告
    print("""
    ### 4.1 Bot 和 Human 类型的数量分布
    **选择原因**:
    通过绘制条形图来展示不同类型的账号数量，可以直观地了解数据集中 bot 和 human 类型账号的比例。

    **结果分析**:
    - 数据集中 `Bot` 类型账号的数量为 {}。
    - 数据集中 `Human` 类型账号的数量为 {}。

    **数据洞察**:
    - 从图表中可以看出，`Human` 类型账号的数量远多于 `Bot` 类型账号，这表明数据集中主要包含的是真实用户的账号。
    """.format(type_counts['Bot'], type_counts['Human']))

    # 4.2 可视化 bot 类型账号的 created_at 情况
    bot_data = data[data['label'] == 'Bot']
    plt.figure(figsize=(10, 6))
    sns.histplot(bot_data['created_at'], bins=30, kde=True)
    plt.title('Bot 类型账号的创建时间分布')
    plt.xlabel('创建时间')
    plt.ylabel('数量')
    plt.show()

    # 报告
    print("""
    ### 4.2 Bot 类型账号的创建时间分布
    **选择原因**:
    通过绘制直方图和核密度估计图来展示 `Bot` 类型账号的创建时间分布，可以了解这些账号的创建时间趋势。

    **结果分析**:
    - `Bot` 类型账号的创建时间集中在 {} 至 {} 之间。
    - 核密度估计图显示，创建时间的分布呈现明显的峰值，表明在某些时间段内 `Bot` 类型账号的创建活动较为频繁。

    **数据洞察**:
    - 从图表中可以看出，`Bot` 类型账号的创建时间分布不均匀，可能与某些特定事件或活动有关。
    """.format(bot_data['created_at'].min(), bot_data['created_at'].max()))

    # 4.3 可视化 human 类型账号的 created_at 情况
    human_data = data[data['label'] == 'Human']
    plt.figure(figsize=(10, 6))
    sns.histplot(human_data['created_at'], bins=30, kde=True)
    plt.title('Human 类型账号的创建时间分布')
    plt.xlabel('创建时间')
    plt.ylabel('数量')
    plt.show()

    # 报告
    print("""
    ### 4.3 Human 类型账号的创建时间分布
    **选择原因**:
    通过绘制直方图和核密度估计图来展示 `Human` 类型账号的创建时间分布，可以了解这些账号的创建时间趋势。

    **结果分析**:
    - `Human` 类型账号的创建时间集中在 {} 至 {} 之间。
    - 核密度估计图显示，创建时间的分布相对平滑，但仍然有一些明显的峰值。

    **数据洞察**:
    - 从图表中可以看出，`Human` 类型账号的创建时间分布较为均匀，但也存在一些高峰期，可能与某些重大事件或活动有关。
    """.format(human_data['created_at'].min(), human_data['created_at'].max()))

    # 4.4 可视化 bot 类型账号的 followers 和 following 情况
    plt.figure(figsize=(12, 6))

    # 绘制 followers 的分布
    plt.subplot(1, 2, 1)
    sns.histplot(bot_data['followers'], bins=30, kde=True)
    plt.title('Bot 类型账号的 followers 分布')
    plt.xlabel('followers 数量')
    plt.ylabel('数量')

    # 绘制 following 的分布
    plt.subplot(1, 2, 2)
    sns.histplot(bot_data['following'], bins=30, kde=True)
    plt.title('Bot 类型账号的 following 分布')
    plt.xlabel('following 数量')
    plt.ylabel('数量')

    plt.tight_layout()
    plt.show()

    # 报告
    print("""
    ### 4.4 Bot 类型账号的 followers 和 following 分布
    **选择原因**:
    通过绘制直方图和核密度估计图来展示 `Bot` 类型账号的 `followers` 和 `following` 分布，可以了解这些账号的社交网络特征。

    **结果分析**:
    - `Bot` 类型账号的 `followers` 数量主要集中在 {} 左右。
    - `Bot` 类型账号的 `following` 数量主要集中在 {} 左右。

    **数据洞察**:
    - 从图表中可以看出，`Bot` 类型账号的 `followers` 和 `following` 数量都相对较低，这可能是因为这些账号的活跃度不高，或者是为了避免被检测而故意限制了社交网络规模。
    """.format(bot_data['followers'].median(), bot_data['following'].median()))

    # 4.5 可视化 human 类型账号的 followers 和 following 情况
    plt.figure(figsize=(12, 6))

    # 绘制 followers 的分布
    plt.subplot(1, 2, 1)
    sns.histplot(human_data['followers'], bins=30, kde=True)
    plt.title('Human 类型账号的 followers 分布')
    plt.xlabel('followers 数量')
    plt.ylabel('数量')

    # 绘制 following 的分布
    plt.subplot(1, 2, 2)
    sns.histplot(human_data['following'], bins=30, kde=True)
    plt.title('Human 类型账号的 following 分布')
    plt.xlabel('following 数量')
    plt.ylabel('数量')

    plt.tight_layout()
    plt.show()

    # 报告
    print("""
    ### 4.5 Human 类型账号的 followers 和 following 分布
    **选择原因**:
    通过绘制直方图和核密度估计图来展示 `Human` 类型账号的 `followers` 和 `following` 分布，可以了解这些账号的社交网络特征。

    **结果分析**:
    - `Human` 类型账号的 `followers` 数量主要集中在 {} 左右。
    - `Human` 类型账号的 `following` 数量主要集中在 {} 左右。

    **数据洞察**:
    - 从图表中可以看出，`Human` 类型账号的 `followers` 和 `following` 数量分布较为广泛，这表明真实用户的社交网络活跃度较高，且个体之间的差异较大。
    """.format(human_data['followers'].median(), human_data['following'].median()))
```
