```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
file_path = r'D:\newcode\dase_introduction\github_bot_raw_data.csv'
raw_data = pd.read_csv(file_path)
print(f"去重前的数据量: {len(raw_data)}")
raw_data.drop_duplicates(inplace=True)
print(f"去重后的数据量: {len(raw_data)}")
raw_data.drop(columns=['gravatar_id'], inplace=True)
print("各列的缺失值情况:")
print(raw_data.isnull().sum())
boolean_columns = ['site_admin', 'company', 'blog', 'location', 'email', 'hireable', 'bio', 'twitter_username']
for col in boolean_columns:
    raw_data[col] = ~raw_data[col].isnull()
text_columns = ['name']
raw_data[text_columns] = raw_data[text_columns].fillna('')
print("处理后的各列的缺失值情况:")
print(raw_data.isnull().sum())
raw_data['created_at'] = pd.to_datetime(raw_data['created_at'])
raw_data['updated_at'] = pd.to_datetime(raw_data['updated_at'])
plt.figure(figsize=(10, 6))
sns.countplot(data=raw_data, x='label')
plt.title('Bot和Human类型的分布')
plt.xlabel('类型')
plt.ylabel('数量')
plt.show()
# 选择原因：柱状图可以直观地展示不同类型的数量对比。
# 结果分析：通过柱状图可以看到Bot和Human的数量差异。
# 数据洞察：如果Bot数量远多于Human，可能需要进一步调查Bot的行为模式。
bot_data = raw_data[raw_data['label'] == 'Bot']
plt.figure(figsize=(10, 6))
sns.histplot(bot_data['created_at'], bins=30, kde=True)
plt.title('Bot类型账号的创建时间分布')
plt.xlabel('创建时间')
plt.ylabel('数量')
plt.show()
# 选择原因：直方图可以展示时间分布的密度。
# 结果分析：通过直方图可以看到Bot账号的创建时间集中在哪些时间段。
# 数据洞察：如果Bot账号的创建时间集中在某个时间段，可能需要调查该时间段内的活动。
human_data = raw_data[raw_data['label'] == 'Human']
plt.figure(figsize=(10, 6))
sns.histplot(human_data['created_at'], bins=30, kde=True)
plt.title('Human类型账号的创建时间分布')
plt.xlabel('创建时间')
plt.ylabel('数量')
plt.show()
# 选择原因：直方图可以展示时间分布的密度。
# 结果分析：通过直方图可以看到Human账号的创建时间集中在哪些时间段。
# 数据洞察：如果Human账号的创建时间集中在某个时间段，可能需要调查该时间段内的活动。
plt.figure(figsize=(10, 6))
sns.scatterplot(data=bot_data, x='followers', y='following')
plt.title('Bot类型账号的followers和following关系')
plt.xlabel('followers')
plt.ylabel('following')
plt.show()
# 选择原因：散点图可以展示两个变量之间的关系。
# 结果分析：通过散点图可以看到Bot账号的followers和following之间的关系。
# 数据洞察：如果大多数Bot账号的followers和following数量都很低，可能表明这些账号活跃度不高。
plt.figure(figsize=(10, 6))
sns.scatterplot(data=human_data, x='followers', y='following')
plt.title('Human类型账号的followers和following关系')
plt.xlabel('followers')
plt.ylabel('following')
plt.show()
# 选择原因：散点图可以展示两个变量之间的关系。
# 结果分析：通过散点图可以看到Human账号的followers和following之间的关系。
# 数据洞察：如果大多数Human账号的followers和following数量较高，可能表明这些账号活跃度较高。
```
